```python
dem = pd.read_csv("C:\\Users\\pc1\\Documents\\Tencent Files\\1064522745\\FileRecv\\a1.csv")
dem.head()
```

```python
# 删去90条Y值为空的数据
dem.Y = dem.Y.apply(lambda x:-1 if pd.isnull(x) else x)
dem = dem[dem.Y>=0]
print len(dem)
```

```python
# dem.EDU_LEVEL.unique()
dem.EDU_LEVEL = dem.EDU_LEVEL.apply(lambda x:-1 if pd.isnull(x) else x)
dem.WORK_PROVINCE = dem.WORK_PROVINCE.apply(lambda x:-1 if pd.isnull(x) else x)
dem.SALARY = dem.SALARY.apply(lambda x:-1 if pd.isnull(x) else x)
dem.AGENT = dem.AGENT.apply(lambda x:-1 if pd.isnull(x) else x)
```

```python
dem['GENDER']=dem['ID_CARD'].apply(lambda x: 1 if int(str(x)[-2])%2!=0 else 0)
dem['BIRTH']=dem['ID_CARD'].apply(lambda x:float(x[:2]))
dem.BIRTH = dem.BIRTH.astype(int)
dem.head()
```

```python
# 对 WP 进行众数回填
keywords=dem.ID_CARD.unique()
for k in keywords:
    temp=dem[dem.ID_CARD==k]
    if (len(temp)==1)&(sum(temp.WORK_PROVINCE)==-1):
        dem.ix[dem.ID_CARD==k,'WORK_PROVINCE']=dem.BIRTH
    if len(temp)>1:
        if sum(temp.WORK_PROVINCE)==(-len(temp)):
            dem.ix[(dem.ID_CARD==k),'WORK_PROVINCE']=dem.BIRTH
        else:
            d=dict(temp.WORK_PROVINCE.value_counts())
            if -1 in d.keys():
                del d[-1]
                dem.ix[(dem.ID_CARD==k)&(dem.WORK_PROVINCE==-1),'WORK_PROVINCE']=max(d.items(), key=lambda x: x[1])[0]
            else:
                continue
            del d
    del temp
dem.head() 
```

```python
def apll1(data1,data2,num):
    data1["Rank"] = data1.Rank.apply(lambda x: num if x in data2 else x)
    return data1
def apll2(data1,data2,num):
    data1["B_Rank"] = data1.B_Rank.apply(lambda x: num if x in data2 else x)
    return data1
gdp1 = gdp[gdp["2016"] >= 20000.0]
gdp2 = gdp[gdp["2016"] >= 10000.0]
gdp2 = gdp2[gdp2["2016"] < 20000.0]
gdp3 = gdp[gdp["2016"] < 10000.0]
gdp_1 = gdp1.code.values
gdp_2 = gdp2.code.values
gdp_3 = gdp3.code.values
dem.WORK_PROVINCE = dem.WORK_PROVINCE.astype(int)
dem["Rank"] = dem.WORK_PROVINCE
dem = apll1(dem,gdp_1,1)
dem = apll1(dem,gdp_2,2)
dem = apll1(dem,gdp_3,3)
print "Rank Finish"
dem.head()
dem.BIRTH = dem.BIRTH.astype(int)
dem["B_Rank"] = dem.BIRTH
dem = apll2(dem,gdp_1,1)
dem = apll2(dem,gdp_2,2)
dem = apll2(dem,gdp_3,3)
print "B_Rank Finish"
dem.head()
```

```python
# 计算 同一ID 出现的次数
cal = dem.ID_CARD.value_counts()
# type(cal)
for k in cal.keys():
    dem.ix[dem.ID_CARD == k,'TIMES'] = cal[k]
#     print k
dem.head()
```

```python
#统计Y_count
ydict={}
keywords=dem.ID_CARD.unique()
for k in keywords:
    temp=dem[dem.ID_CARD==k]
    ydict[k]=sum(temp.Y)
    del temp
for k in keywords:
    dem.ix[dem.ID_CARD==k,'Y_counts']=ydict[k]
```

```python
#还款日期
import time
import datetime
def getdaysdiff(stra,strb):
    ta=time.strptime(stra,'%Y/%m/%d')
    tb=time.strptime(strb,'%Y/%m/%d')
    y,m,d= ta[0:3]
    dataTimea=datetime.datetime(y,m,d)
    y,m,d = tb[0:3]
    dataTimeb=datetime.datetime(y,m,d)
    daysdiff=abs((dataTimea-dataTimeb).days)
    return daysdiff
keywords=dem[dem.TIMES>1].ID_CARD.unique()
kdict={}
for k in keywords:
    mindiff=365
    temp=dem[dem.ID_CARD==k]
    timestring=temp.LOAN_DATE.values
    length=len(timestring)
    for i in range(length):
        for j in range(i+1,length):
            x=getdaysdiff(timestring[i],timestring[j])
            if x<mindiff:
                mindiff=x
    kdict[k]=mindiff
for k in keywords:
    dem.ix[dem.ID_CARD==k,'TIMEDIFF']=kdict[k]
```

```python
#EDU_COUNTS,MARYY_STATUS_COUNTS
keywords=dem.ID_CARD.unique()
for k in keywords:
    temp=dem[dem.ID_CARD==k]
    dem.ix[dem.ID_CARD==k,'EDU_counts']=len(temp.EDU_LEVEL.unique())
    dem.ix[dem.ID_CARD==k,'MARRY_counts']=len(temp.MARRY_STATUS.unique())
    dem.ix[dem.ID_CARD==k,'WORK_PROVINCE_counts']=len(temp.WORK_PROVINCE.unique())
    del temp
dem.head()
```

```python
#IS_CHEAT构造
times=dem.TIMES.values
edu=dem.EDU_counts.values
marry=dem.MARRY_counts.values
timediff=dem.TIMEDIFF.values
work=dem.WORK_PROVINCE_counts.values
ischeat=[]
length=len(times)
for i in range(length):
    temp=0
    if times[i]==1:
        ischeat.append(0)
        continue
    else:
        if marry[i]>2:
            temp+=1
        if edu[i]>2:
            temp+=1
        if timediff[i]<=7:
            temp+=1
        if work[i]>2:
            temp+=1
        ischeat.append(temp)
dem['IS_CHEAT']=ischeat
```
