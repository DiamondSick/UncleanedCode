```python
import pandas as pd
import numpy as np
import time
import datetime
```

```python
df=pd.read_csv("F:\\contest\\MCM\\a4.csv")
```

```python
#删除没有标签的数据
df['Y'] = df['Y'].apply(lambda x: -1 if pd.isnull(x) else x)
df=df[df.Y>=0]
df.drop('REPORT_ID',axis=1,inplace=True)
```

```python
#统计记录的times
cal=df['ID_CARD'].value_counts()
for k in cal.keys():
    df.ix[df.ID_CARD==k,'times']=cal[k]
df.head()
```

```python
#空值填-1
df['MARRY_STATUS']=df['MARRY_STATUS'].apply(lambda x:-1 if pd.isnull(x) else x)
df['WORK_PROVINCE']=df['WORK_PROVINCE'].apply(lambda x:-1 if pd.isnull(x) else x)
df['EDU_LEVEL']=df['EDU_LEVEL'].apply(lambda x:-1 if pd.isnull(x) else x)
df['SALARY']=df['SALARY'].apply(lambda x:-1 if pd.isnull(x) else x)
df['AGENT']=df['AGENT'].apply(lambda x:-1 if pd.isnull(x) else x)
```

```python
#移数据（老师的失误）
df.ix[df.EDU_LEVEL==4,'MARRY_STATUS']=0
df.ix[df.EDU_LEVEL==6,'MARRY_STATUS']=3
df.ix[df.EDU_LEVEL==9,'MARRY_STATUS']=4
df.ix[df.EDU_LEVEL==4,'EDU_LEVEL']=-1
df.ix[df.EDU_LEVEL==6,'EDU_LEVEL']=-1
df.ix[df.EDU_LEVEL==9,'EDU_LEVEL']=-1
df.ix[df.EDU_LEVEL==5,'EDU_LEVEL']=4
df.ix[df.EDU_LEVEL==7,'EDU_LEVEL']=5
df.ix[df.EDU_LEVEL==8,'EDU_LEVEL']=6
df.ix[df.EDU_LEVEL==10,'EDU_LEVEL']=7
df.ix[df.EDU_LEVEL==11,'EDU_LEVEL']=8
```

```python
#从ID_CARD中提取性别和出生地 1男0 女
df['GENDER']=df['ID_CARD'].apply(lambda x: 1 if int(str(x)[-2])%2!=0 else 0)
df['BIRTH']=df['ID_CARD'].apply(lambda x:float(x[:2]))
```

```python
#统计Y
ydict={}
keywords=df.ID_CARD.unique()
for k in keywords:
    temp=df[df.ID_CARD==k]
    ydict[k]=sum(temp.Y)
    del temp
for k in keywords:
    df.ix[df.ID_CARD==k,'Y_counts']=ydict[k]
#y的频率
den=df['Y_counts'].values
dum=df['times'].values
result=[den[i]/dum[i] for i in range(len(den))]
df['Y_fre']=result
```

```python
#还款日期
def getdaysdiff(stra,strb):
    ta=time.strptime(stra,'%Y/%m/%d')
    tb=time.strptime(strb,'%Y/%m/%d')
    y,m,d= ta[0:3]
    dataTimea=datetime.datetime(y,m,d)
    y,m,d = tb[0:3]
    dataTimeb=datetime.datetime(y,m,d)
    daysdiff=abs((dataTimea-dataTimeb).days)
    return daysdiff
keywords=df[df.times>1].ID_CARD.unique()
kdict={}
for k in keywords:
    mindiff=365
    temp=df[df.ID_CARD==k]
    timestring=temp.LOAN_DATE.values
    length=len(timestring)
    for i in range(length):
        for j in range(i+1,length):
            x=getdaysdiff(timestring[i],timestring[j])
            if x<mindiff:
                mindiff=x
    kdict[k]=mindiff
for k in keywords:
    df.ix[df.ID_CARD==k,'timediff']=kdict[k]
```

```python
#EDU_COUNTS,MARYY_STATUS_COUNTS,WORK_PROVINCE_COUNTS
keywords=df.ID_CARD.unique()
for k in keywords:
    temp=df[df.ID_CARD==k]
    df.ix[df.ID_CARD==k,'EDU_counts']=len(temp.EDU_LEVEL.unique())
    df.ix[df.ID_CARD==k,'MARRY_counts']=len(temp.MARRY_STATUS.unique())
    df.ix[df.ID_CARD==k,'WORK_PROVINCE_counts']=len(temp.WORK_PROVINCE.unique())
    del temp
```

```python
#AGENT标签替换
df.ix[df.AGENT=='wechat','AGENT']=3
df.ix[(df.AGENT=='kuaiqian')|(df.AGENT=='weijinhui')|(df.AGENT=='orgloan')|(df.AGENT=='bestpay'),'AGENT']=0
df.ix[(df.AGENT=='APP')|(df.AGENT=='ifpp')|(df.AGENT=='fenqile')|(df.AGENT=='rongzhijia'),'AGENT']=1
df.ix[(df.AGENT=='chinapnr')|(df.AGENT=='huifusdb'),'AGENT']=2
```

```python
#IS_CHEAT构造
times=df.times.values
edu=df.EDU_counts.values
marry=df.MARRY_counts.values
timediff=df.timediff.values
work=df.WORK_PROVINCE_counts.values
ischeat=[]
length=len(times)
for i in range(length):
    temp=0
    if times[i]==1:
        ischeat.append(0)
        continue
    else:
        if marry[i]>2:
            temp+=1
        if edu[i]>2:
            temp+=1
        if timediff[i]<=7:
            temp+=1
        if work[i]>2:
            temp+=1
        ischeat.append(temp)
df['IS_CHEAT']=ischeat
```

```python
df.to_csv("F:\\contest\\MCM\\raw.csv",index=None)
```
